Now, we have 2,592 documents that comprise our corpus. From here, we can clean and process our text. 

```python
def process_text_nltk(
    text: str,
    lower_case: bool = True,
    remove_punct: bool = True,
    remove_stopwords: bool = False,
    lemma: bool = False,
    string_or_list: str = "str",
):
    tokens = nltk.word_tokenize(text)

    if lower_case:
        tokens = [token.lower() if token.isalpha() else token for token in tokens]

    if remove_punct:
        tokens = [token for token in tokens if token.isalpha()]

    if remove_stopwords:
        tokens = [token for token in tokens if not token in STOPLIST]

    if lemma:
        tokens = [nltk.wordnet.WordNetLemmatizer().lemmatize(token) for token in tokens]

    if string_or_list != "list":
        doc = " ".join(tokens)
    else:
        doc = tokens

    return doc
```

```python
df["tokens_stopwords"] = df.text.apply(process_text_nltk, remove_stopwords = False, string_or_list = "list")
df["tokens_no_stopwords"] = df.text.apply(process_text_nltk, remove_stopwords = True, string_or_list = "list")
```

```python
docs_list_stopwords = list(df.tokens_stopwords)
docs_list_no_stopwords = list(df.tokens_no_stopwords)
```
