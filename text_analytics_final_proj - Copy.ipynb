Now, we have 2,592 documents that comprise our corpus. From here, we can clean and process our text. 

```python
def process_text_nltk(
    text: str,
    lower_case: bool = True,
    remove_punct: bool = True,
    remove_stopwords: bool = False,
    lemma: bool = False,
    string_or_list: str = "str",
):
    tokens = nltk.word_tokenize(text)

    if lower_case:
        tokens = [token.lower() if token.isalpha() else token for token in tokens]

    if remove_punct:
        tokens = [token for token in tokens if token.isalpha()]

    if remove_stopwords:
        tokens = [token for token in tokens if not token in STOPLIST]

    if lemma:
        tokens = [nltk.wordnet.WordNetLemmatizer().lemmatize(token) for token in tokens]

    if string_or_list != "list":
        doc = " ".join(tokens)
    else:
        doc = tokens

    return doc
```

```python
df["tokens_stopwords"] = df.text.apply(process_text_nltk, remove_stopwords = False, string_or_list = "list")
df["tokens_no_stopwords"] = df.text.apply(process_text_nltk, remove_stopwords = True, string_or_list = "list")
df["text2"] = df.text.apply(process_text_nltk, remove_stopwords = True, string_or_list = "str")
```

```python
# quinn removal
def remove_quinn(tokens):
    return [token for token in tokens if token != 'quinn' and token != 'maeve' and token != 'cassius' and token != 'ty' and token != 'val' and token != 'valerie']

# Remove 'quinn' from each list in the 'tokens' column
df['tokens_stopwords'] = df['tokens_stopwords'].apply(remove_quinn)
df['tokens_no_stopwords'] = df['tokens_no_stopwords'].apply(remove_quinn)
```

```python
docs_list_stopwords = list(df.tokens_stopwords)
docs_list_no_stopwords = list(df.tokens_no_stopwords)
```

```python
from gensim.models import phrases
from gensim.models import Phrases
```

```python
bigram_model = Phrases(docs_list_stopwords, min_count=1, delimiter = "_", connector_words = phrases.ENGLISH_CONNECTOR_WORDS)

# Print bigram in document
for token in bigram_model[docs_list_stopwords[11]]:
    if "_" in token:
        print(token)
```

```python
def replace_bigrams_with_phrases(token_list):
    return bigram_model[token_list]

df["tokens_bigrams"] = df.tokens_no_stopwords.apply(replace_bigrams_with_phrases)
docs_list = list(df.tokens_bigrams)
```

```python
trigram_model = Phrases(docs_list, min_count=10, delimiter = "_", connector_words = phrases.ENGLISH_CONNECTOR_WORDS)

# Print bigram in document
for token in trigram_model[docs_list[2]]:
    if "_" in token:
        print(token)
```

```python
def replace_trigrams_with_phrases(token_list):
    return trigram_model[token_list]
df["tokens_trigrams"] = df.tokens_bigrams.apply(replace_trigrams_with_phrases)
docs_list = list(df.tokens_trigrams)
```

Now, we have a column of tokens, stopword-free tokens, tokens with bigrams, and tokens with trigrams. Let's build LDA models on all four.

```python
gensim_dictionary_stopwords = Dictionary(df.tokens_stopwords)
gensim_dictionary_stopwords.filter_extremes(no_below=10, no_above=0.65)

gensim_dictionary_no_stopwords = Dictionary(df.tokens_no_stopwords)
gensim_dictionary_no_stopwords.filter_extremes(no_below=10, no_above=0.65)

gensim_dictionary_bigrams = Dictionary(df.tokens_bigrams)
gensim_dictionary_bigrams.filter_extremes(no_below=10, no_above=0.65)

gensim_dictionary_trigrams = Dictionary(df.tokens_trigrams)
gensim_dictionary_trigrams.filter_extremes(no_below=10, no_above=0.65)
```

```python
corpus_stopwords = []
for doc in df.tokens_stopwords:
    corpus_stopwords.append(gensim_dictionary_stopwords.doc2bow(doc))

corpus_no_stopwords = []
for doc in df.tokens_no_stopwords:
    corpus_no_stopwords.append(gensim_dictionary_no_stopwords.doc2bow(doc))

corpus_bigrams = []
for doc in df.tokens_bigrams:
    corpus_bigrams.append(gensim_dictionary_bigrams.doc2bow(doc))

corpus_trigrams = []
for doc in df.tokens_trigrams:
    corpus_trigrams.append(gensim_dictionary_trigrams.doc2bow(doc))
```

```python
NUM_TOPICS = 50
RANDOM_STATE = 5643

temp = gensim_dictionary_stopwords[0] # required to initialize, see what happens if you comment out
word_id = gensim_dictionary_stopwords.id2token

lda_stopwords_50 = LdaModel(
    corpus=corpus_stopwords,
    id2word=word_id,
    random_state = RANDOM_STATE,
    num_topics=NUM_TOPICS,
)

temp = gensim_dictionary_no_stopwords[0] 
word_id = gensim_dictionary_no_stopwords.id2token

lda_no_stopwords_50 = LdaModel(
    corpus=corpus_no_stopwords,
    id2word=word_id,
    random_state = RANDOM_STATE,
    num_topics=NUM_TOPICS,
)

temp = gensim_dictionary_bigrams[0] 
word_id = gensim_dictionary_bigrams.id2token

lda_bigrams_50 = LdaModel(
    corpus=corpus_bigrams,
    id2word=word_id,
    random_state = RANDOM_STATE,
    num_topics=NUM_TOPICS,
)

temp = gensim_dictionary_trigrams[0] 
word_id = gensim_dictionary_trigrams.id2token

lda_trigrams_50 = LdaModel(
    corpus=corpus_trigrams,
    id2word=word_id,
    random_state = RANDOM_STATE,
    num_topics=NUM_TOPICS,
)
```
